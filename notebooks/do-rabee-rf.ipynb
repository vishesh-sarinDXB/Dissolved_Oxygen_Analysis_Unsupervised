```r
library(pheatmap)
library(data.table)
```

```r
do_data = data.frame(fread("do_data_anot.csv"))
do_data = do_data[-2]
do_data = do_data[-2]
do_data = do_data[-2]
do_data = do_data[-2]
do_data = do_data[-7]
do_data = do_data[-8]
row.names(do_data) = do_data[, 1]
do_data = do_data[-1]

do_data_month_medians = data.frame(fread("do_data__month_med.csv"))
do_data_month_medians = (do_data_month_medians[-9])
do_data_month_medians = (do_data_month_medians[-2])
do_data_month_medians = (do_data_month_medians[-2])
do_data_month_medians = do_data_month_medians[-1]
row.names(do_data_month_medians) = do_data_month_medians[, 6]
do_data_month_medians = do_data_month_medians[-6]

do_data_tb_medians = data.frame(fread("do_data_tb_medians.csv"))
do_data_tb_medians = do_data_tb_medians[-1]
row.names(do_data_tb_medians) = do_data_tb_medians[, 6]
do_data_tb_medians = do_data_tb_medians[-6]

do_data_mtb_medians = data.frame(fread("do_data_mtb_medians.csv"))
row.names(do_data_mtb_medians) = do_data_mtb_medians[, 1]
do_data_mtb_medians = do_data_mtb_medians[-1]
do_data_mtb_medians = do_data_mtb_medians[-1]
do_data_mtb_medians = do_data_mtb_medians[-1]
do_data_mtb_medians = do_data_mtb_medians[-1]
```
